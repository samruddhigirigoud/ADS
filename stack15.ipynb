```python
class StackNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class StackWithMiddle:
    def __init__(self):
        self.head = None
        self.mid = None
        self.size = 0
    
    def push(self, value):
        new_node = StackNode(value)
        if self.head is None:
            self.head = new_node
            self.mid = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            if self.size % 2 == 0:
                self.mid = self.mid.prev
        self.size += 1
    
    def pop(self):
        if self.head is None:
            return None
        popped_value = self.head.value
        self.head = self.head.next
        if self.head:
            self.head.prev = None
        if self.size % 2 == 1:
            self.mid = self.mid.next
        self.size -= 1
        return popped_value
    
    def get_middle(self):
        if self.mid is None:
            return None
        return self.mid.value
    
    def display(self):
        temp = self.head
        stack_values = []
        while temp:
            stack_values.append(temp.value)
            temp = temp.next
        print("Stack:", stack_values)
        print("Middle Element:", self.get_middle())

# Example usage
stack = StackWithMiddle()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

print("Stack after pushes:")
stack.display()

print("Popped Element:", stack.pop())
print("Stack after pop:")
stack.display()

print("Middle Element:", stack.get_middle())
```
