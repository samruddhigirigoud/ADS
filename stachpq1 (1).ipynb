```python
def reverse_string_using_stack(s):
    stack = []
    for char in s:
        stack.append(char)  # Push each character onto the stack
    reversed_str = "".join(stack.pop() for _ in range(len(stack)))  # Pop and reverse
    return reversed_str

# Example usage
input_str = "Hello"
print(reverse_string_using_stack(input_str))  # Output: "olleH"
```

```python
class Stack:
    def __init__(self):
        self.stack = []
    
    def push(self, value):
        self.stack.append(value)
    
    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)
    
    def sorted_insert(self, value):
        if self.is_empty() or value > self.peek():
            self.push(value)
        else:
            temp = self.pop()
            self.sorted_insert(value)
            self.push(temp)
    
    def sort_stack(self):
        if not self.is_empty():
            temp = self.pop()
            self.sort_stack()
            self.sorted_insert(temp)
    
    def display(self):
        print(self.stack)

# Example usage
stack = Stack()
stack.push(3)
stack.push(1)
stack.push(4)
stack.push(2)
stack.push(5)

print("Original stack:")
stack.display()

stack.sort_stack()

print("Sorted stack:")
stack.display()
```

```python
class Stack:
    def __init__(self):
        self.stack = []
    
    def push(self, value):
        self.stack.append(value)
    
    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)

class UndoRedo:
    def __init__(self):
        self.undo_stack = Stack()
        self.redo_stack = Stack()
    
    def perform_action(self, action):
        self.undo_stack.push(action)
        self.redo_stack = Stack()  # Clear redo stack after new action
    
    def undo(self):
        if not self.undo_stack.is_empty():
            action = self.undo_stack.pop()
            self.redo_stack.push(action)
            return action
        return None
    
    def redo(self):
        if not self.redo_stack.is_empty():
            action = self.redo_stack.pop()
            self.undo_stack.push(action)
            return action
        return None
    
    def display(self):
        print("Undo Stack:", self.undo_stack.stack)
        print("Redo Stack:", self.redo_stack.stack)

# Example usage
undo_redo = UndoRedo()
undo_redo.perform_action("Action 1")
undo_redo.perform_action("Action 2")
undo_redo.perform_action("Action 3")

print("After actions:")
undo_redo.display()

print("Undo:", undo_redo.undo())
print("Undo:", undo_redo.undo())
undo_redo.display()

print("Redo:", undo_redo.redo())
undo_redo.display()
```

```python
class StackNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class StackWithMiddle:
    def __init__(self):
        self.head = None
        self.mid = None
        self.size = 0
    
    def push(self, value):
        new_node = StackNode(value)
        if self.head is None:
            self.head = new_node
            self.mid = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            if self.size % 2 == 0:
                self.mid = self.mid.prev
        self.size += 1
    
    def pop(self):
        if self.head is None:
            return None
        popped_value = self.head.value
        self.head = self.head.next
        if self.head:
            self.head.prev = None
        if self.size % 2 == 1:
            self.mid = self.mid.next
        self.size -= 1
        return popped_value
    
    def get_middle(self):
        if self.mid is None:
            return None
        return self.mid.value
    
    def display(self):
        temp = self.head
        stack_values = []
        while temp:
            stack_values.append(temp.value)
            temp = temp.next
        print("Stack:", stack_values)
        print("Middle Element:", self.get_middle())

# Example usage
stack = StackWithMiddle()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

print("Stack after pushes:")
stack.display()

print("Popped Element:", stack.pop())
print("Stack after pop:")
stack.display()

print("Middle Element:", stack.get_middle())
```

```python
class Stack:
    def __init__(self):
        self.stack = []
    
    def push(self, value):
        self.stack.append(value)
    
    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)
    
    def remove_duplicates(self):
        unique_elements = set()
        new_stack = []
        while not self.is_empty():
            element = self.pop()
            if element not in unique_elements:
                unique_elements.add(element)
                new_stack.append(element)
        while new_stack:
            self.push(new_stack.pop())
    
    def display(self):
        print("Stack:", self.stack)

# Example usage
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(2)
stack.push(4)
stack.push(3)
stack.push(5)

print("Stack before removing duplicates:")
stack.display()

stack.remove_duplicates()

print("Stack after removing duplicates:")
stack.display()
```

```python
from collections import deque

class StackUsingQueues:
    def __init__(self):
        self.queue1 = deque()
        self.queue2 = deque()
    
    def push(self, value):
        self.queue1.append(value)
    
    def pop(self):
        if self.is_empty():
            return None
        while len(self.queue1) > 1:
            self.queue2.append(self.queue1.popleft())
        popped_value = self.queue1.popleft()
        self.queue1, self.queue2 = self.queue2, self.queue1  # Swap queues
        return popped_value
    
    def peek(self):
        if self.is_empty():
            return None
        while len(self.queue1) > 1:
            self.queue2.append(self.queue1.popleft())
        top_value = self.queue1.popleft()
        self.queue2.append(top_value)
        self.queue1, self.queue2 = self.queue2, self.queue1  # Swap queues
        return top_value
    
    def is_empty(self):
        return len(self.queue1) == 0
    
    def display(self):
        print("Stack:", list(self.queue1))

# Example usage
stack = StackUsingQueues()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

print("Stack after pushes:")
stack.display()

print("Popped Element:", stack.pop())
print("Stack after pop:")
stack.display()

print("Top Element:", stack.peek())
```

```python
class Stack:
    def __init__(self):
        self.stack = []
    
    def push(self, value):
        self.stack.append(value)
    
    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return None
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)

def next_greater_elements(arr):
    stack = Stack()
    result = [-1] * len(arr)
    
    for i in range(len(arr) - 1, -1, -1):
        while not stack.is_empty() and stack.peek() <= arr[i]:
            stack.pop()
        if not stack.is_empty():
            result[i] = stack.peek()
        stack.push(arr[i])
    
    return result

# Example usage
arr = [4, 5, 2, 10, 8]
print("Input List:", arr)
print("Next Greater Elements:", next_greater_elements(arr))
```
